```python
import pandas as pd
import numpy as np
import pandas_ta as ta
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Thuchanh/FPT.csv',index_col=0)
```

```python
data.head()
```

```python
data.isnull()
```

```python
data.dropna()
```

# MA AND RSI


```python
data['MA50'] = data['Close'].rolling(window=50).mean()
```

```python
data['RSI']= ta.rsi(data['Close'],length=14)
```

```python
data
#check data NaN
```

```python
# tiền dữ liệu lần 2
data.fillna(data['MA50'].mean(),inplace=True)
data.fillna(data['RSI'].mean(),inplace=True)
#nhược điểm khi kết thúc mô hình sẽ có thể ảnh hương đến kết quả dự đoán thay vì vậy nên xóa hết các cột NaN nhưng vì để toàn vẹn dữ liệu 
#nên sủ dụng phương thức fill giá trị trung bình
```

```python
train, test = train_test_split(data,test_size=0.2,shuffle=False)
```

```python
data.fillna(data['MA50'].mean(),inplace=True)
data.fillna(data['RSI'].mean(),inplace=True)
train[['MA50','RSI']].isnull().sum()
```

# Data visualization
 

```python
plt.figure(figsize=(12, 6))
plt.bar(data.index, data['Volume'], color='blue')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Trading Volume Over Time')
plt.show()
```

# Train Model

```python
model = LinearRegression()
```

```python
model
```

```python
imputer = SimpleImputer(strategy='mean')
imputer.fit(train[['MA50', 'RSI']]) 

#thay đổi chiến lược huấn luyện học máy vi các vấn đề vì dữ liệu bị lỗi
```

```python
X_train_imputed = imputer.transform(train[['MA50', 'RSI']])
y_train = train['Close'] 
```

```python
model.fit(X_train_imputed, y_train)
```

```python
predictions = model.predict(test[['MA50','RSI']])
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test['Close'], predictions)
print(f'Mean Squared Error: {mse}')
```

```python
future_data = data[['MA50', 'RSI']].copy()
future_predictions = model.predict(future_data)
```
